### 1. Basic Plotting

First things first, we need to `import matplotlib.pyplot` to access the plotting functions.

#### 1.1 Creating Simple Line Plots
As the name suggests, data points are connected by straight lines which are useful for displaying data that varies continuously over a range, making it easy to identify patterns and trends. Use `plt.plot(x,y)` for a simple line plot.

**Note:** Older Python code and some environments may require the command `plt.show()` to display the graph. This is no longer necessary in many environments (including Positron), and is omitted in the code below. 

````python
import matplotlib.pyplot as plt

# Let's take a data representing monthly website traffic (in thousands)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
traffic = [150, 200, 180, 220, 250, 210]

# To create a line plot
plt.plot(months, traffic)
````

But as we can see, it doesn’t have any labels or titles as such.

#### 1.2 Customization options for labels, colors, and styles.

We can convey the information with much clarity by customizing the plots. Matplotlib offers numerous customization options, allowing you to control color, line style, markers, and more.

* Add Labels and Title: Use `plt.xlabel()`, `plt.ylabel()` and `plt.title()`

````python
# Sample data representing monthly website traffic (in thousands)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
traffic = [150, 200, 180, 220, 250, 210]

# To create a line plot
plt.plot(months, traffic)

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Website Traffic')
````

* Changing color and linestyle: 
Use the parameter marker to mark the points, linestyle to change the styling of line, and add grid to the plots by using plt.grid(True).

````python
# Create a line plot with custom appearance
plt.plot(months, traffic, marker='o', linestyle='--', color='g')

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Monthly Traffic (in Thousands)')
plt.title('Monthly Website Traffic')

# Add grid to the plot
plt.grid(True)
````

**Note:** See this guide (https://codepointtech.com/customize-matplotlib-plots/) for more customization options.

To change the figure size: We often have to adjust the plot size, right? And to do that, we need to use plt.figure(figsize=(x_size,y_size)). Make sure to use this before the .plot command.


* Using Label: 
In the case of plotting multiple lines in the same graph, you can do so by using the plot command two times for the variables you want. But the issue is to differentiate them properly, for this, we have a parameter called a label, along with that you also need to use plt.legend().

#### Saving Matplotlib plots as image files.

Matplotlib enables you to save the plot as an image file.

* In Jupyter Notebook: 
When working in Jupyter Notebook if you wish to save the plot as an image file, you have to use plt.savefig('path/to/directory/plot_name.png') . You can specify the complete file path, and you can specify the desired file name and format( Eg: .jpg, .png, .pdf )

````python
# Sample data for two products' monthly revenue (in thousands of dollars)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
product_a_revenue = [45, 55, 60, 70, 80, 75]
product_b_revenue = [35, 40, 50, 55, 70, 68]

# Create a line plot for Product A with a blue line and circular markers
plt.plot(months, product_a_revenue, marker='o', linestyle='-', color='blue', label='Product A')

# Create a line plot for Product B with a red dashed line and square markers
plt.plot(months, product_b_revenue, marker='s', linestyle='--', color='red', label='Product B')

# Add labels and a title
plt.xlabel('Month')
plt.ylabel('Monthly Revenue (in $1000)')
plt.title('Monthly Revenue Comparison')

# Display a legend to differentiate between Product A and Product B
plt.legend()

# Save the plot as png
plt.savefig('my_plot.png')

# Save the plot as jpg
plt.savefig('my_plot.jpg')
````

#### 2.1 Bar Chart

Bar charts represent categorical data with rectangular bars, where the length or height of each bar represents a value. You can use the command `plt.bar(x,y)` to generate vertical bar charts and `plt.barh(x,y)` for horizontal bar charts.

Use Case:

* Comparing sales performance of different products.
* Showing population distribution by country.

````python
# bring in numpy for data manipulation
import numpy as np

# Define the categories and expenses for 3 people
categories = ['Rent', 'Groceries', 'Utilities', 'Transport', 'Entertainment']
person1 = [1200, 600, 200, 300, 150]
person2 = [1000, 700, 300, 200, 100]
person3 = [1100, 500, 250, 400, 200]

# Calculate the positions for each bar
barWidth = 0.25  # Width of the bars
r1 = np.arange(len(categories))  # Positions for person1's bars
r2 = [x + barWidth for x in r1]  # Shift person2's bars by barWidth
r3 = [x + barWidth for x in r2]  # Shift person3's bars by barWidth

# Create the bar chart
plt.figure(figsize=(10,6))

# Plotting the bars for each person
plt.bar(r1, person1, color='brown', width=barWidth, edgecolor='grey', label='Person 1')
plt.bar(r2, person2, color='black', width=barWidth, edgecolor='grey', label='Person 2')
plt.bar(r3, person3, color='green', width=barWidth, edgecolor='grey', label='Person 3')

# Add labels and title
plt.xlabel('Categories', fontweight='bold')  # X-axis label
plt.ylabel('Expenses ($)', fontweight='bold')  # Y-axis label
plt.xticks([r + barWidth for r in range(len(categories))], categories)  # Set category names as x-ticks
plt.title('Monthly Expense Comparison for 3 People')  # Chart title

# Create legend & Show plot
plt.legend()
````

#### 2.2 Histograms

* Histograms are used to visualize the distribution of continuous or numerical data and they help us identify patterns in data. 
* In a histogram plot the data is grouped into “bins,” and the height of each bar represents the frequency or count of data points in that bin. 
* It takes the lower and upper limits of the given data and divides it into the number of bins given.

You can use the command `plt.hist(x)` to generate histograms. Unlike the bar plot, here you don’t need the ‘y’, as it only represents the frequency of one continuous data. The default bins are 10, and they can be changed. You can override the bins range as well with your desired bins range. You can also add the edgecolor for bars.

Along with the histogram plot, in the same graph if you want to add any line, say the mean or median, you can do so by calculating the value and passing to to plt.axvline(calculated_mean,label=desired_label). This can be used with any other plot.

Use Cases:
* Analyzing age distribution in a population.
* Examining exam score distribution in a classroom.

````python
# Sample exam scores data
exam_scores = [68, 72, 75, 80, 82, 84, 86, 90, 92, 95, 98, 100]

# Custom bin ranges
bin_ranges = [60, 70, 80, 90, 100]

# Create a histogram with custom bin ranges by assigning it to bins.
plt.hist(exam_scores, bins=bin_ranges, color='lightblue', edgecolor='black', alpha=0.7)

# Add labels and a title
plt.xlabel('Exam Scores')
plt.ylabel('Frequency')
plt.title('Exam Scores Histogram with Custom Bins')

# Calculate and add a median line
median_score = np.median(exam_scores)
plt.axvline(median_score, color='red', linestyle='dashed', linewidth=2, label=f'Median Score: {median_score}')

# Add a legend
plt.legend()
````

#### 2.3 Scatter Plots

Scatter plots are graphs used to visualize the relationship between two variables. Each point on the plot corresponds to a single data point and represents the values of these two variables. 

* Scatter plots are graphs that use dots to represent the values obtained for two different variables, one plotted along the x-axis and the other plotted along the y-axis.
* They show the spread of data points and can indicate the relationship (correlation) between the variables.

Use Cases:
* To identify relationships between variables: Helps in understanding if two variables are correlated.
* To spot outliers or anomalies: Useful for detecting data points that deviate significantly from the rest.

````python
# Generating sample data
x = np.random.randn(100)
y = x * np.random.randn(100)

# Creating scatter plot
plt.scatter(x, y, color='blue', alpha=0.5)  # Plot points with a blue color and semi-transparency

# Customizing the plot
plt.title('Scatter Plot')
plt.xlabel('X values')
plt.ylabel('Y values')
````

#### 2.4 Pie Charts

Pie charts represent parts of a whole as slices of a circular pie. 
They are suitable for showing the composition of a single categorical variable in terms of percentages. But this won’t look good when there are more than six categories as they get clumsy, in such cases horizontal bar might be preferred. Bar charts do better at precisely indicating the magnitudes of each category, whereas pie charts show rough proportions. It is not recommended to shadow or change the viewing angle of the pie, since this interferes with accurate perception of the proportions.

* Use the command plt.pie(x,labels=your_category_names, colors=desired_colors_list) , if you have a desired colors list, you can provide that and you can also change the edge color of the pie chart with the parameter wedgeprops={'edgecolor':your_color} .

We can also highlight particular segments using explode parameter by passing a tuple where each element is the amount by which it has to explode. And autopct parameter enables you to choose how many values after the decimal are to be shown in the plot.

Use Cases:

* Displaying the distribution of a budget by expense categories.
* Showing the market share of various smartphone brands.

Eg: Exploding a particular segment for better storytelling during presentations.



````python
# Product categories
categories = ['Electronics', 'Clothing', 'Home Decor', 'Books', 'Toys']

# Sales data for each category
sales = [3500, 2800, 2000, 1500, 1200]

# Explode a specific segment (e.g., 'Clothing'),
# second value (0.1) is the amount by which the segment 'Clothing'.
explode = (0, 0.1, 0, 0, 0)  

# Create a pie chart with explode and shadow
plt.pie(sales, labels=categories, explode=explode, shadow=False, autopct='%1.1f%%')
plt.title('Sales by Product Category')
````

#### 2.5 Box Plot (Box and Whisker Plot)

Box plots summarizes the distribution of numerical data by displaying quartiles, outliers, and potential skewness. They provide insights into data spread, central tendency, and variability. Box plots are especially useful for identifying outliers and comparing distributions.

You can use plt.boxplot(data) to plot the box plot. You can customize the appearance of the box and outliers using boxprops and flierprops , use vert=False to make the box plot horizontal and patch_artist=True to fill the box with color.

Use Cases:
* Analyzing the distribution of salaries in a company.
* Assessing the variability of housing prices in different neighborhoods.

````python
# Generating sample data for three groups
np.random.seed(16)  # For reproducible results
group1 = np.random.normal(100, 10, 200)
group2 = np.random.normal(90, 20, 200)
group3 = np.random.normal(80, 30, 200)

data = [group1, group2, group3]  # Combine data into a list

# Creating the box plot
fig, ax = plt.subplots()
bp = ax.boxplot(data, patch_artist=True, notch=True, tick_labels=['Group 1', 'Group 2', 'Group 3'])

# Customizing the plot
ax.set_title('Box Plot')
ax.set_ylabel('Values')
ax.set_xlabel('Groups')

# Adding colors
colors = ['lightblue', 'lightgreen', 'tan']
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
````

#### 2.6 Heatmap

Heatmaps are graphical representations of data where individual values contained in a matrix are represented as colors. Color-coded to show different values, offering an immediate visual summary of information.

Use Cases:
* To represent the magnitude of phenomena: Useful in various fields, such as showing geographical data or statistical data.
* To visualize correlation matrices: Highlights the strength of the relationship between variables.

`plt.imshow()` is a Matplotlib function that is used for displaying 2D image data, visualizing 2D arrays, or showing images in various formats.

Using imshow for heatmap: Here, we’ll create a heatmap to visualize a correlation matrix, and we’ll use a color map to show this relationship visually. Pass the correlation matrix to imshow to visualize the heatmap.

````python
# Create a sample correlation matrix
correlation_matrix = np.array([[1.0, 0.8, 0.3, -0.2],
                               [0.8, 1.0, 0.5, 0.1],
                               [0.3, 0.5, 1.0, -0.4],
                               [-0.2, 0.1, -0.4, 1.0]])

# Create a heatmap for the correlation matrix
plt.imshow(correlation_matrix, cmap='coolwarm', vmin=-1, vmax=1, aspect='auto', origin='upper')

# Add a colorbar with the following commands
cbar = plt.colorbar()
cbar.set_label('Correlation', rotation=270, labelpad=20)

# Add labels and a title
plt.title('Correlation Matrix Heatmap')
plt.xticks(range(len(correlation_matrix)), ['Var1', 'Var2', 'Var3', 'Var4'])
plt.yticks(range(len(correlation_matrix)), ['Var1', 'Var2', 'Var3', 'Var4'])

# note using plt.show() does remove the sometimes messy extra information in the output
plt.show()
````

#### 2.7 Stack Plot

Imagine you want to visualize how three product categories (electronics, clothing, and home appliances) contribute to total sales over four quarters (Q1 to Q4). Then you can represent each category’s sales as layers in the plot, and the plot helps us understand their contributions and trends over time. That’s exactly what the stack plot does.

A stack plot, which is also known as a stacked area plot, is a type of data visualization that displays multiple datasets as layers stacked on top of one another, with each layer representing a different category or component of the data. Stack plots are particularly useful for visualizing how individual components contribute to a whole over a continuous time period or categorical domain. Use it as plt.stackplot(x,y1,y2) , as many stacks as you want!

````python
# Sample data for stack plot
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
electronics = [10000, 12000, 11000, 10500]
clothing = [5000, 6000, 7500, 8000]
home_appliances = [7000, 7500, 8200, 9000]

# Create a stack plot
plt.figure(figsize=(10, 6))  # Set the figure size
plt.stackplot(quarters, electronics, clothing, home_appliances, labels=['Electronics', 'Clothing', 'Home Appliances'],
              colors=['blue', 'green', 'red'], alpha=0.7)

# Add labels, legend, and title
plt.xlabel('Quarters')
plt.ylabel('Sales ($)')
plt.title('Product Category Sales Over Quarters')
plt.legend(loc='lower right')

# Display the plot
plt.grid(True)
````

### 3. Multiple Subplots

#### 3.1 Creating Multiple Plots in a Single Figure

If you are working with a dataset that has the age of a person, the software they are working on, and their salary. You want to visualize the Python developers' ages and salaries and then compare them with Java developers. By now, you know you can do that by making two plots one in each cell of the notebook. But then, you have to toggle back and forth to compare, we better not talk about what if there are 4 things to compare!

To ease this issue, we have a feature called subplots, in the same plot there will be different subplots of each. You can create the subplots using plt.subplots(nrows=x,ncols=y). By default nrows=1, and ncols=1. plt.subplots() returns two things one(fig) is to style the entire plot, and the other(axes) is to make subplots. Plot each subplot using axes[row, column], where row and column specify the location of the subplot in the grid. You can use the sharex or sharey parameters when you have common axes for the subplots.

````python
# Sample data
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(nrows=2,ncols=2, figsize=(10, 8))

# Plot the first subplot (top-left) - axes[0,0]
axes[0, 0].plot(x, y1, color='blue')
axes[0, 0].set_title('Sine Function')

# Plot the second subplot (top-right) - axes[0,1]
axes[0, 1].plot(x, y2, color='green')
axes[0, 1].set_title('Cosine Function')

# Plot the third subplot (bottom-left) - axes[1,0]
axes[1, 0].plot(x, y3, color='red')
axes[1, 0].set_title('Tangent Function')

# Plot the fourth subplot (bottom-right) - axes[1,1]
axes[1, 1].plot(x, y4, color='purple')
axes[1, 1].set_title('Exponential Function')

# Adjust spacing between subplots
plt.tight_layout()

# Add a common title for all subplots
fig.suptitle('Various Functions', fontsize=16)
````

####  3.2 Combining Different Types of Plots

When talking about comparing plots, we will not always wish to have the same axes for both plots, right? There will be cases where we have one common axis and other different!

In such cases, You can combine these different plots within a single figure using the twinx() or twiny() functions to share one axis while having independent y or x-axes. For example, you can combine a line plot of Month vs. Revenue, with a bar plot of Month vs. Sales, to visualize two related datasets with different scales. Here we have a common x-axis but a different y-axis.

````python
# Sample data
x = np.arange(1, 6)
y1 = np.array([10, 15, 7, 12, 9])
y2 = np.array([200, 300, 150, 250, 180])

# Create a bar plot
fig, ax1 = plt.subplots(figsize=(8, 4))
ax1.bar(x, y1, color='b', alpha=0.7, label='Sales')
ax1.set_xlabel('Month')
ax1.set_ylabel('Sales', color='b')
ax1.set_ylim(0, 20)  # Set y-axis limits for the left y-axis

# Create a line plot sharing the x-axis
ax2 = ax1.twinx()
ax2.plot(x, y2, color='r', marker='o', label='Revenue')
ax2.set_ylabel('Revenue', color='r')
ax2.set_ylim(0, 400)  # Set y-axis limits for the right y-axis

# Add a legend
fig.legend(loc='upper left', bbox_to_anchor=(0.15, 0.85))

# Add a title
plt.title('Sales and Revenue Comparison')

# Show the plot
plt.show()
````

## II. Plotly

### What is Plotly?
Plotly is a powerful open-source graphing library that enables the creation of interactive, publication-quality visualizations in a wide range of programming languages, including Python. It allows users to create complex visualizations such as line charts, scatter plots, bar charts, 3D graphs, and more, all with the ability to zoom, hover, and interact with the data.

Plotly supports both **Plotly Express** (a high-level interface for creating easy visualizations) and the **Plotly Graph Objects API** (a low-level API that offers more control and customization).

### Importance of Interactive Visualization
Interactive visualizations are essential in today's data-driven world because they:
- Enhance user engagement by allowing real-time data exploration.
- Make it easier to identify patterns and insights by allowing dynamic interactions like zooming, filtering, and tooltips.
- Provide flexibility for large datasets, enabling users to focus on specific aspects of the data without overwhelming them with information.
- Improve presentations and storytelling, making data communication more effective and memorable.

### Why Plotly for Data Visualization?
Plotly is a preferred tool for data visualization due to:
- **Interactivity**: Plotly visualizations are highly interactive by default, providing a better way to explore data.
- **Ease of Use**: Plotly Express allows for simple, concise creation of complex charts with minimal code.
- **Customization**: The Plotly Graph Objects API offers advanced customization for those who need more control over visual details.
- **Wide Range of Visualizations**: From basic charts (line, bar, scatter) to advanced types (3D plots, geographic maps, financial charts), Plotly covers nearly every visualization need.
- **Cross-Language Support**: Plotly works in Python, R, Julia, and JavaScript, making it versatile for different platforms.
- **Web-Based Sharing**: Plotly visualizations can be embedded in web applications or dashboards easily, enhancing collaboration.

### Overview of Plotly Features and Capabilities
- **Interactivity**: Zoom, pan, hover, and click events make it easy to explore data.
- **3D Plotting**: Supports 3D visualizations like scatter, surface, and mesh plots.
- **Mapping**: With built-in support for geographic maps, Plotly can easily integrate with `GeoJSON` for dynamic mapping.
- **Animations**: Supports animated charts that reflect changes in data over time.
- **Custom Layouts**: Offers complete control over plot layouts, titles, axes, legends, and annotations.
- **Integration**: Works seamlessly with dashboards (e.g., Dash) and other tools like Jupyter Notebooks, making it ideal for data science workflows.

Choosing between Plotly and Matplotlib depends on several factors related to the specific needs of the visualization task at hand. Here are some considerations to help you decide when to prefer one library over the other:

### When to Prefer Matplotlib
- **Static Visualizations**: When you need simple, static charts, and interactivity isn't necessary (e.g., for reports, papers).
- **Speed**: Matplotlib is faster for generating basic plots, especially with large datasets.
- **Customization**: While Plotly allows for customization, Matplotlib's low-level API offers more detailed control for fine-tuning every aspect of the plot.
- **Minimal Dependency**: If interactivity isn’t needed and you want a lightweight solution with fewer dependencies.

### When to Prefer Plotly over Matplotlib
- **Interactivity is Essential**: Plotly should be preferred when the goal is to allow users to interact with visualizations by zooming, hovering, or filtering.
- **Advanced Visualizations**: Plotly excels at advanced chart types like 3D plots, geographic maps, and animations, which require complex interactivity.
- **Web-Based Applications**: Plotly is better suited for web-based or real-time interactive dashboards, particularly when using Plotly Dash for web deployment.
- **Ease of Embedding**: Plotly makes it straightforward to embed interactive charts in web apps or shareable platforms, which is more cumbersome in Matplotlib.

### When to Use Both
There are scenarios where combining **Matplotlib** and **Plotly** can be beneficial:
- **Static and Interactive Hybrid Reports**: Use Matplotlib for static charts in printed reports and Plotly for interactive elements in online versions.
- **Complex Layouts**: You can use Matplotlib to create specialized, multi-panel static plots, and Plotly for individual interactive charts that require dynamic exploration.
- **Specialized Use Cases**: If Matplotlib's static visualizations are combined with Plotly's dynamic features in dashboards or notebooks, users get the best of both worlds, allowing for easy switching between static and interactive modes.

In practice, many data analysts and scientists switch between Plotly and Matplotlib depending on the task.

## Plotly Charts: Integrating Plotly with Pandas for Data Visualization

Plotly can be easily integrated with **Pandas** to create interactive visualizations directly from your DataFrame.

### 1. **Line Plot**
A line plot displays data as a series of points connected by straight line segments. It is useful for visualizing trends over time or continuous data.

````python
import plotly.express as px  # Importing Plotly Express, a high-level library for data visualization
import pandas as pd  # Importing Pandas for data manipulation

# Creating a DataFrame with sample data using Pandas.
# The DataFrame contains two columns: 'Year' and 'Sales'.
df = pd.DataFrame({
    'Year': [2015, 2016, 2017, 2018, 2019],  # The 'Year' column represents the years
    'Sales': [200, 250, 300, 350, 400]       # The 'Sales' column represents the sales figures for each year
})

# Creating a line chart using Plotly Express.
# The 'x' axis is set to the 'Year' column, and the 'y' axis is set to the 'Sales' column.
# The 'title' parameter is used to set the chart's title to 'Sales Over Years'.
fig = px.line(df, x='Year', y='Sales', title='Sales Over Years')

# Display the plot in an interactive format.

fig.show()  # This renders the chart in the notebook or interactive environment
df
````

### 2. Bar Plot

A bar plot displays categorical data with rectangular bars. Each bar's length or height is proportional to the values it represents.

````python
# Create a bar chart using Plotly
fig = px.bar(df, x='Year', y='Sales', title='Yearly Sales')

# Show the plot
fig.show()
````

### 3. Scatter Plot

A scatter plot uses dots to represent the values of two different variables. It helps to show the relationship or correlation between the variables.

````python
# Create a scatter plot using Plotly
fig = px.scatter(df, x='Year', y='Sales', title='Sales Scatter Plot')

# Show the plot
fig.show()
````

### 4. Pie Chart

A pie chart shows proportions by dividing a circle into slices, where each slice represents a category's share of the whole.

### 7. Animated Chart

Animated charts help visualize changes in data over time by adding animation frames to the charts.

````python
import plotly.graph_objects as go  # Importing Plotly Graph Objects for more customizable plots
import numpy as np  # Importing NumPy for numerical computations

# Generate sample data for temperatures across months and years
np.random.seed(42)  # Setting a random seed for reproducibility
years = range(1990, 2021)  # Defining the years from 1990 to 2021
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']  # Defining the months

# Creating a list to store the generated temperature data
data = []
for year in years:
    base_temp = np.random.normal(15, 5)  # Generate a base temperature for the year (mean=15, std=5)
    for month in months:
        # Calculate temperature by adding seasonal variation (sin wave) and noise (mean=0, std=2)
        temp = base_temp + np.random.normal(0, 2) + 5 * np.sin(np.pi * months.index(month) / 6)
        # Append the data for each month and year
        data.append({'Year': year, 'Month': month, 'Temperature': temp})

# Convert the list of dictionaries into a Pandas DataFrame
df = pd.DataFrame(data)

# Create a figure object to contain the plot
fig = go.Figure()

# Add the initial heatmap plot to the figure
# Heatmap uses Month (x-axis), Year (y-axis), and Temperature (color-coded)
heatmap = go.Heatmap(
    x=df['Month'],  # X-axis represents months
    y=df['Year'],   # Y-axis represents years
    z=df['Temperature'],  # Z-axis represents temperature values (color intensity)
    colorscale='RdBu_r',  # Color scale representing temperature (red-blue reversed)
    zmin=df['Temperature'].min(),  # Minimum temperature for color scaling
    zmax=df['Temperature'].max()   # Maximum temperature for color scaling
)
# Add the heatmap trace to the figure
fig.add_trace(heatmap)

# Create frames for the animation, one frame for each year
frames = []
for year in years:
    # Each frame updates the heatmap, showing data for up to the current year
    frame = go.Frame(
        data=[go.Heatmap(
            z=df[df['Year'] <= year]['Temperature'],  # Temperature data for all years up to the current year
            zmin=df['Temperature'].min(),  # Maintain consistent color scaling
            zmax=df['Temperature'].max()
        )],
        name=str(year)  # Set the frame name to the current year for slider control
    )
    frames.append(frame)

# Assign the frames to the figure for the animation
fig.frames = frames

# Add a slider to control the animation frame by year
sliders = [dict(
    active=0,  # Start the slider at the first frame (year 1990)
    currentvalue={"prefix": "Year: "},  # Display the current year on the slider
    pad={"t": 50},  # Padding to position the slider
    steps=[dict(
        label=str(year),  # Label each step with the year
        method="animate",  # Link each step to the 'animate' method
        args=[[str(year)], dict(frame=dict(duration=300, redraw=True), transition=dict(duration=300))]  # Frame transition duration
    ) for year in years]  # Create a slider step for each year
)]

# Add Play and Pause buttons for animation control
updatemenus = [dict(
    type="buttons",
    buttons=[dict(
        label="Play",  # Play button starts the animation
        method="animate",  # Use the 'animate' method to play the frames
        args=[None, {"frame": {"duration": 300, "redraw": True},  # Frame duration for smooth transition
                     "fromcurrent": True,  # Continue animation from the current frame
                     "transition": {"duration": 300, "easing": "quadratic-in-out"}}]),  # Smooth easing for transitions
             dict(
        label="Pause",  # Pause button stops the animation
        method="animate",
        args=[[None], {"frame": {"duration": 0, "redraw": False},  # Stop the animation immediately
                       "mode": "immediate",  # Pause without transitions
                       "transition": {"duration": 0}}])  # No transition when pausing
    ],
    direction="left",  # Position the buttons to the left
    pad={"r": 10, "t": 87},  # Padding to position the buttons
    showactive=False,  # Hide active button highlighting
    x=0.1,  # X-position of the buttons
    xanchor="right",  # Anchor to the right
    y=0,  # Y-position of the buttons
    yanchor="top"  # Anchor to the top
)]

# Update the layout of the figure to include the title, axis labels, slider, and buttons
fig.update_layout(
    title="Temperature Changes Over Years",  # Title of the chart
    xaxis_title="Month",  # X-axis label
    yaxis_title="Year",  # Y-axis label
    sliders=sliders,  # Add the slider for year control
    updatemenus=updatemenus  # Add the play/pause buttons
)

# Show the figure with the heatmap and animation controls
fig.show()
````

### 4. **Customizing Legends**
The legend explains what each color or line represents in your chart. You can customize its position, font, and orientation.

````python
import numpy as np

# Generate sample data
# np.linspace creates an array of 100 equally spaced numbers from 0 to 10
x = np.linspace(0, 10, 100)

# Create a DataFrame with x values and trigonometric functions
df = pd.DataFrame({
    'x': x,
    'Sin': np.sin(x),  # Sine function
    'Cos': np.cos(x),  # Cosine function
    'Tan': np.tan(x)   # Tangent function
})

# Create a line plot using plotly express
# px.line automatically creates separate traces for each y column
fig = px.line(df,               # Input DataFrame
              x='x',            # Column to use for x-axis
              y=['Sin', 'Cos', 'Tan'],  # Columns to use for y-axis (multiple lines)
              title='Trigonometric Functions')

# Customize the layout of the plot
fig.update_layout(
    xaxis_title='X',  # Set x-axis title
    yaxis_title='Y',  # Set y-axis title
    
    # Customize the legend
    legend=dict(
        # Position the legend in the top-right corner of the plot
        x=1,           # 1 corresponds to the right edge of the plot
        y=0,           # 1 corresponds to the bottom edge of the plot
        xanchor='right',  # Anchor point is on the right of the legend box
        yanchor='bottom',    # Anchor point is at the bottom of the legend box
        
        # Set the orientation of the legend
        orientation='v',  # 'v' for vertical, 'h' for horizontal
        
        # Customize the font of the legend text
        font=dict(
            family='Arial, sans-serif',  # Font family
            size=12,                     # Font size
            color='black'                # Font color
        ),
        
        # Add a border around the legend
        bordercolor='black',  # Color of the border
        borderwidth=1,        # Width of the border in pixels
        
        # Set the background color of the legend
        bgcolor='white',      # White background
        
        # Adjust the spacing between legend items
        itemsizing='constant',  # Keeps sizes of legend items constant
        
        # Add a title to the legend
        title='Functions',    # Text of the legend title
        title_font_family='Times New Roman, Times, serif'  # Font for the legend title
    )
)

# Display the plot
fig.show()
````

### 5. **Customizing Grid**
Gridlines can help read values from the plot more easily. You can control their appearance, including color, width, and visibility.

````python
import plotly.graph_objects as go

# Generate sample data
np.random.seed(42)  # Set a seed for reproducibility so that the random numbers generated are the same every time
x = np.random.randn(100)  # Generate 100 random points from a standard normal distribution for x-values
y = 2 * x + np.random.randn(100) * 0.5  # Create y-values based on a linear relationship (2 * x), with added random noise

# Create the scatter plot using Plotly
fig = go.Figure(data=go.Scatter(
    x=x,                       # x-values for the scatter plot
    y=y,                       # y-values for the scatter plot
    mode='markers',            # Set mode to 'markers' to create a scatter plot
    marker=dict(
        size=8,                # Set the marker size to 8
        color=y,               # Set the marker color based on the y-values (color gradient)
        colorscale='Viridis',  # Use the 'Viridis' colorscale for the markers (from dark to light colors)
        showscale=True         # Show the color scale bar on the right side of the plot
    )
))

# Customize the layout, including gridlines and axis settings
fig.update_layout(
    title='Customized Grid Example',    # Set the title of the plot

    # X-axis customization
    xaxis=dict(
        title='X-axis',                # Set the X-axis title
        showgrid=True,                 # Display gridlines for the X-axis
        gridcolor='lightgray',         # Set the color of the X-axis gridlines to light gray
        gridwidth=1,                   # Set the width of the X-axis gridlines
        zeroline=True,                 # Display a line at x=0 (zeroline)
        zerolinecolor='red',           # Set the color of the X-axis zeroline to red
        zerolinewidth=2,               # Set the width of the zeroline to make it more visible
    ),

    # Y-axis customization
    yaxis=dict(
        title='Y-axis',                # Set the Y-axis title
        showgrid=True,                 # Display gridlines for the Y-axis
        gridcolor='lightblue',         # Set the color of the Y-axis gridlines to light blue
        gridwidth=1,                   # Set the width of the Y-axis gridlines
        zeroline=True,                 # Display a line at y=0 (zeroline)
        zerolinecolor='red',           # Set the color of the Y-axis zeroline to red
        zerolinewidth=2,               # Set the width of the zeroline to make it more prominent
        griddash='dash',               # Make the Y-axis gridlines dashed for better distinction
    ),
    
    # Background customization
    plot_bgcolor='white',              # Set the background color of the plot to white for a clean look
)

# Add minor gridlines to the X-axis for more precision in reading values
fig.update_xaxes(
    minor=dict(
        showgrid=True,                # Enable minor gridlines on the X-axis
        gridcolor='lightgray',        # Set the color of minor X-axis gridlines to light gray
        gridwidth=0.5,                # Set the width of minor X-axis gridlines to be thinner
        griddash='dot',               # Make the minor gridlines dotted to differentiate from the major gridlines
    )
)

# Add minor gridlines to the Y-axis for more precision in reading values
fig.update_yaxes(
    minor=dict(
        showgrid=True,                # Enable minor gridlines on the Y-axis
        gridcolor='lightblue',        # Set the color of minor Y-axis gridlines to light blue
        gridwidth=0.5,                # Set the width of minor Y-axis gridlines to be thinner
        griddash='dot',               # Make the minor gridlines dotted to differentiate from the major gridlines
    )
)

# Display the plot
fig.show()  # Renders the plot with all customizations, including gridlines, axis settings, and background color
````

### Bonus Section: ShinyLive and Plotly Integration

**ShinyLive** is a platform that enables the creation of interactive web applications using Python or R directly in the browser. It eliminates the need for backend servers, making it easy to deploy and share apps that use dynamic visualizations.

In ShinyLive, the Python or R code runs directly in the user's browser through WebAssembly, which allows high-performance, interactive apps without requiring any server-side processing. This is particularly useful for building apps that provide interactive data visualization, dashboards, and analytical tools, all while being lightweight and easy to distribute.

#### Key Features of ShinyLive:
- **No Backend Required**: Everything runs in the browser, meaning no server setup or deployment is needed.
- **Supports Python/R**: You can use popular data analysis and visualization libraries in your app.
- **Interactive and Responsive**: ShinyLive apps are highly interactive and respond quickly to user inputs such as slider adjustments or button clicks.
- **Easy to Share**: Since there’s no backend, you can easily share your app by providing a URL, embedding it in a website, or sharing the code.

#### Uses of Plotly on ShinyLive:
In ShinyLive, you can integrate Plotly to enhance the interactivity of your web apps by allowing users to interact with dynamic charts and graphs.

**Some uses of Plotly in ShinyLive include:**

- **Interactive Dashboards**: Combining Plotly with ShinyLive allows the creation of interactive dashboards where users can interact with the data directly through Plotly charts. They can zoom, pan, hover over data points, and more.

- **Customizable Visualizations**: With Plotly, you can offer customizable chart options, such as allowing users to change the axes, modify chart types, or update data dynamically, all within the ShinyLive environment.

- **Enhanced User Engagement**: The ability to interact with charts, visualize complex datasets, and receive instant feedback on inputs makes Plotly an excellent tool for engaging users in analytical tasks and data exploration.

### Steps to Run ShinyLive App:

1. **Create a Python file (app.py):**
   - Write your ShinyLive app code in a file called `app.py`.

2. **Navigate to the directory:**
   - Use the `cd` command to change the directory to where your `app.py` file is located:
     ```bash
     cd /path/to/your/app/directory
     ```

3. **Run the ShinyLive app:**
   - Run the app on port 8080 using the following command:
     ```bash
     shiny run app.py --port 8080
     ```

4. **Access the app:**
   - Once the app is running, a URL will be displayed in the terminal. Copy and paste the URL into your browser to access the app.

Example URL: `http://127.0.0.1:8080`
